### 1. Countries with most content

```python
countries = list(data["country"].dropna())
countries
```

```python
country_dict = {}

for country in countries:
    for c in country.split(", "):
        if c in country_dict:
            country_dict[c] += 1
        else:
            country_dict[c] = 1
```

```python
sorted_country = sorted(country_dict.items(), key = lambda x: x[1], reverse=True)  # Sort by value
sorted_country[0][0]
```

```python
pd.DataFrame(sorted_country,columns=["Country","Number of contents"]).head(10)
```

### 2. Content types by Directors

```python
director_data = data[["director","listed_in"]].dropna()
director_data
```

```python
director_list = director_data["director"]
director_dict = {}
for directors in director_list:
    for director in directors.split(", "):
        director_dict[director] = {}
```

```python
for row in director_data.values: # Get director and genre
    for director in director_dict:  # Get director's name
        if director in row[0]:  # Check if the director match with director_data
            for types in row[1:]:  # Get all genres
                for type_ in types.split(", "):  # Loop each genre
                    if type_ in director_dict[director]:
                        director_dict[director][type_] += 1
                    else:
                        director_dict[director][type_] = 1
director_dict
```

```python
for director in director_dict:
    genre = sorted(director_dict[director].items(), key = lambda x: x[1], reverse=True)[0][0]
    director_dict[director] = genre
pd.DataFrame(director_dict.items(),columns=["Director","Genre"])
```

### 10. Most used words for title

```python
titles = list(data["title"].dropna())
titles
```

```python
from nltk.corpus import stopwords

all_stopwords = stopwords.words('english')
```

```python
import re

title_dict = {}
for title in titles:
    for word in title.split(" "):
        word = re.sub('[^a-zA-Z]','',word)
        word = word.lower()
        if word not in all_stopwords and word != "":
            if word.lower() in title_dict:
                title_dict[word.lower()] += 1
            else:
                title_dict[word.lower()] = 1
```

```python
sorted_title = sorted(title_dict.items(), key = lambda x: x[1], reverse=True)  # Sort by value
pd.DataFrame(sorted_title,columns=["Word","Frequency"]).head(15)
```
